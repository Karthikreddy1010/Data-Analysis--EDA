```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
"""Questions
1) Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?
2) How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity?
3) How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?
4) Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?
5) How do different fruit varieties (Variety) differ in terms of size, weight, sweetness, acidity, and ripeness?"""
```

```python
data=pd.read_csv("Orange Quality Data.csv")
data.head()
```

```python
print("Total rows in a dataset:",len(data))
data=data.drop_duplicates()
print("After removing the duplicates rows in a dataset:",len(data))
```

# 1) Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?


```python
size_weight=data[['Size (cm)','Weight (g)']].corr()
print("Correlation of Orange Size and Weight:",size_weight)
```

```python
sns.heatmap(size_weight,annot=True)
```

```python
sns.scatterplot(data,x='Size (cm)', y='Weight (g)',size='Size (cm)')
plt.title("Size will Correlate with Weight of Oranges")
```

###  The size of the orange increases,its weight tends to increase as well, but the correlation is not very strong.

# 2) How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity?


```python
data.head(1)
```

```python
sweet_acidic=data[['Brix (Sweetness)','pH (Acidity)']].corr()
print("Correlation of Sweetness and PH:",sweet_acidic)
```

```python
sns.scatterplot(data=data,x='Brix (Sweetness)',y='pH (Acidity)')
plt.title("Orange Sweetnes will correlate with PH acidity")
```

```python
sweet_ph=data[['Brix (Sweetness)','pH (Acidity)']]
print("Sweetness and PH acidity of oranges data:",sweet_ph)
```

```python
sns.pairplot(sweet_ph)
```

```python
sns.heatmap(sweet_acidic,annot=True)
```

### Higher pH values indicate lower acidity,so sweeter oranges may be slightly less acidic.
### The relationship between sweetness and acidity is weak, so other factors likely contribute to the sweetness or acidity of oranges

# 3) How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?


```python
data.head(1)
```

```python
softness=data[['Softness (1-5)','Ripeness (1-5)']].corr()
print("The correlation of Softness and Ripeness of Oranges:",softness)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Softness (1-5)'], kde=True, color='orange', label='Softness', stat="density", alpha=0.5)
sns.kdeplot(data['Ripeness (1-5)'], color='blue', label='Ripeness', lw=2)
plt.title("Softness vs Ripeness Distribution with KDE Line")
plt.xlabel("Softness (1-5)")
plt.ylabel("Ripeness (1-5)")
plt.legend()
plt.show()
```

### The softness of the orange increases, the ripeness tends to decrease slightly,

# 4) Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?


```python
data.head(1)
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='Blemishes (Y/N)', y='Ripeness (1-5)', data=data,hue='Blemishes (Y/N)')
plt.title('Ripeness Levels by Blemish Presence')
plt.xticks(rotation=45)  
plt.show()

plt.figure(figsize=(14, 6))
sns.boxplot(x='Blemishes (Y/N)', y='Softness (1-5)', data=data, hue='Blemishes (Y/N)')
plt.title('Softness Levels by Blemish Presence')
plt.xticks(rotation=45)  
plt.show()
```

# 5) How do different fruit varieties (Variety) differ in terms of size, weight, sweetness, acidity, and ripeness?"""

```python
data.head(1)
```

```python
data=data[['Size (cm)','Weight (g)','Brix (Sweetness)','pH (Acidity)','Ripeness (1-5)','Variety']]
data.head(1)
```

```python
sns.pairplot(data,hue='Variety',height=3,diag_kind='kde')
plt.suptitle("Orange Varieties will differ in size, weight,sweetness,acidity, and Ripeness",y=1.02,fontsize=16)
plt.tight_layout()
plt.show()
```

```python
grouped_data = data.groupby('Variety')[['Size (cm)','Weight (g)','Brix (Sweetness)','pH (Acidity)','Ripeness (1-5)']].mean()

plt.figure(figsize=(10, 6))
sns.heatmap(grouped_data, annot=True, cmap='coolwarm')
plt.title("Mean Feature Values by Variety")
plt.show()
```
